```python
import pandas as pd
import numpy as np
corr_matrix = df_num.corr()
np.fill_diagonal(corr_matrix.values, np.nan)
max_corr = corr_matrix.stack().idxmax()  
max_value = corr_matrix.stack().max()

print(f"The highest correlation is between {max_corr[0]} and {max_corr[1]} with a value of {max_value:.2f}")
```
